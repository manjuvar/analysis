```python
results_json = response.json()
results_item = results_json['search_results']
results_item[0].keys()
```

<font color='yellow'>Put everything together - Loop through results and append data inside a list</font>

```python
model = []
mileage = []
year = []
dealer_name = []
price = []

for result in results_item:

    #model
    model.append(result['model'])
    #mileage
    mileage.append(result['mileage'])
    #year
    year.append(result['year'])
    #dealer name
    dealer_name.append(result['dealer_name'])
    #price
    price.append(result['price'])
```

<font color='yellow'> Pandas Dataframe</font>

```python
tesla_df = pd.DataFrame({'Model':model, 'Mileage':mileage, 'Year':year, 
                         'Dealer_name': dealer_name, 'Price':price})

tesla_df
```

<font color='yellow'>Scraping multiple pages  </font>

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64; rv:109.0) Gecko/20100101 Firefox/112.0',
    'Accept': '*/*',
    'Accept-Language': 'en-US,en;q=0.5',
    # 'Accept-Encoding': 'gzip, deflate, br',
    'Referer': 'https://www.autolist.com/',
    'Origin': 'https://www.autolist.com',
    'DNT': '1',
    'Connection': 'keep-alive',
    'Sec-Fetch-Dest': 'empty',
    'Sec-Fetch-Mode': 'cors',
    'Sec-Fetch-Site': 'same-site',
}

model = []
mileage = []
year = []
dealer_name = []
price = []

for i in range(1,6):

    params = {
    'city': 'San Francisco',
    'latitude': '37.7749295',
    'limit': '20',
    'longitude': '-122.4194155',
    'make': 'Tesla',
    'page': str(i),
    'radius': '50',
    'state': 'CA',
}

    # response variable for the get request
    response = requests.get('https://production-proxy.autolist.com/api/cwv/seo/listings', params=params, headers=headers)

    #json object
    results_json = response.json()

    #result items (20 items for each page)
    results_item = results_json['search_results']

    for result in results_item:

        #model
        model.append(result['model'])
        #mileage
        mileage.append(result['mileage'])
        #year
        year.append(result['year'])
        #dealer names
        dealer_name.append(result['dealer_name'])
        #price
        price.append(result['price'])
```

```python
tesla_multiple_df = pd.DataFrame({'Model':model, 'Mileage':mileage, 'Year':year, 
                         'Dealer_name': dealer_name, 'Price':price})
```
