```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.select import Select
import time
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
firefox_profile = webdriver.FirefoxProfile()
firefox_profile.set_preference('browser.privatebrowsing.autostart', True)
driver = webdriver.Firefox(firefox_profile=firefox_profile)
driver.get('https://imdb.com')
#driver.maximize_window()

#dropdown
dropdown = driver.find_element(By.CLASS_NAME , 'ipc-icon--arrow-drop-down')
dropdown.click()
time.sleep(1)

#advanced search from dropdown
element = driver.find_element(By.LINK_TEXT, 'Advanced Search')
element.click()

#click on advanced title search
adv_title = driver.find_element(By.LINK_TEXT, 'Advanced Title Search')
adv_title.click()

#select feature film
feature_film = driver.find_element(By.ID, 'title_type-1')
feature_film.click()

#select feature film
tv_movie = driver.find_element(By.ID, 'title_type-2')
tv_movie.click()

#min release date
min_date = driver.find_element(By.NAME, 'release_date-min')
min_date.click()
min_date.send_keys('1990')

#max release date
max_date = driver.find_element(By.NAME, 'release_date-max')
max_date.click()
max_date.send_keys('2020')

#user rating_min
rating_min = driver.find_element(By.NAME, 'user_rating-min')
rating_min.click()
dropdown_2 = Select(rating_min)
dropdown_2.select_by_visible_text('1.0')


#user rating_max
rating_max = driver.find_element(By.NAME, 'user_rating-max')
rating_max.click()
dropdown_3 = Select(rating_max)
dropdown_3.select_by_visible_text('10')

#oscar nominated
oscar_nom = driver.find_element(By.ID, 'groups-7')
oscar_nom.click()

#color
color = driver.find_element(By.ID, 'colors-1')
color.click()

#language
language = driver.find_element(By.NAME, 'languages')
dropdown_4 = Select(language)
dropdown_4.select_by_visible_text('English')

#250 results
no_of_pages = driver.find_element(By.ID, 'search-count')
dropdown_5 = Select(no_of_pages)
dropdown_5.select_by_index(2)

#search button
search = driver.find_element(By.XPATH, "(//button[@type='submit'])[2]")
search.click()

#current
current_url = driver.current_url

#######################

#get request
response = requests.get(current_url)

#soup object
soup = BeautifulSoup(response.text, 'html.parser')

#result items(starting point)
list_items = soup.find_all('div', {'class':'lister-item'})

#list comprehension
movie_title = [list_item.find('h3', {'class': 'lister-item-header'}).a.text for list_item in list_items]
year = [list_item.find('span', {'class': 'lister-item-year'}).text.
        replace('(','').replace(')','') for list_item in list_items]
duration = [list_item.find('span', {'class': 'runtime'}).text for list_item in list_items]
genre = [list_item.find('span', {'class': 'genre'}).text.strip()  for list_item in list_items]
rating = [list_item.find('div', {'class': 'inline-block ratings-imdb-rating'}).text.strip()
           for list_item in list_items]

#create dataframe
imdb_df = pd.DataFrame({'Title': movie_title, 'Year': year, 'Duration': duration, 'Rating':rating})

#output in excel
#imdb_df.to_excel('imdb_multiple_pages.xlsx', index=False)
```
